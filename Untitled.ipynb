# Importando Dados

```python
import pandas as pd
```

```python

dados = pd.read_csv('tips.csv')
```

```python
dados.head()
```

# Tradução

```python
dados.columns
```

```python
renomear = {
    'total_bill' : 'valor_da_conta',
    'tip' : 'gorjeta',
    'dessert' : 'sobremesa',
    'day' : 'dia_da_semana',
    'time' : 'hora_do_dia',
    'size' : 'total_de_pessoas'
}
```

```python
gorjetas = dados.rename(columns=renomear)
```

```python
gorjetas.head(1)
```

```python
gorjetas.sobremesa.unique()
```

```python
sim_nao = {
    'No' : 'Não',
    'Yes' : 'Sim'
}
```

```python
gorjetas.sobremesa = gorjetas.sobremesa.map(sim_nao)
```

```python
gorjetas.dia_da_semana.unique()
```

```python
dias = {
    'Sun' : 'Domingo',
    'Sat' : 'Sabado',
    'Thur': 'Quinta',
    'Fri' : 'Sexta'
}
gorjetas.dia_da_semana = gorjetas.dia_da_semana.map(dias)
```

```python
gorjetas.head(1)
```

```python
gorjetas.hora_do_dia.unique()
```

```python
hora = {
    'Dinner' : 'Jantar', 
    'Lunch' : 'Almoço'
}

gorjetas.hora_do_dia = gorjetas.hora_do_dia.map(hora)
```

```python
gorjetas.head(1)
```

## Teste de hipótese

**H<sup>null</sup>**
a distribuição da taxa das gorjetas é a mesma nos dois grupos

**H<sup>alt</sup>**
a distribuição da taxa das gorjetas não é a mesma nos dois grupos

```python
from scipy.stats import ranksums
```

```python
sobremesa = gorjetas.query('sobremesa == "Sim"').porcentagem
sem_sobremesa = gorjetas.query('sobremesa == "Não"').porcentagem
```

```python
r = ranksums(sobremesa, sem_sobremesa)
print("O valor de p-value é {}".format(r.pvalue))
```

**H<sup>null</sup>**
a distribuição da taxa das gorjetas é a mesma nos dois grupos
